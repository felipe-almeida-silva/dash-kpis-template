```python
df = pd.read_csv('old_sales.csv')
```

```python
columns = df.columns.to_list()
```

```python
# lower case columns
columns = [x.lower() for x in columns]
```

```python
# replace space with _ and - with _
columns = [x.replace(" ", "_") for x in columns]
columns = [x.replace("-", "_") for x in columns]
```

```python
# convert Order Date and Ship Date to date '%Y-%m-%d'
df['Order Date'] = pd.to_datetime(df['Order Date']).dt.strftime('%Y-%m-%d')
df['Ship Date'] = pd.to_datetime(df['Ship Date']).dt.strftime('%Y-%m-%d')
```

```python
# drop nan
df = df.dropna()
```
